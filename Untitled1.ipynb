```python
import numpy as np
import sympy
import matplotlib.pyplot as plt
import scipy.interpolate as inter
import numpy.polynomial.chebyshev as che #Использовался NumPy v1.14
from mpmath import *
```

```python
def f(x):
    return (1 / (1 + 25*x**2))
n = 15
x = np.linspace(-1, 1, n)
t = np.linspace (-1, 1, 101)
```

```python
'''Строим интерполяцию многочленом Лагранжа'''

for k in [5, 9, 19]:
    xk = np.linspace(-1, 1, k)
    l = inter.lagrange(xk, f(xk))
    plt.plot(t, l(t))

plt.grid(True)
plt.plot(x,f(x),'k.')

plt.show()
```

```python
'''Теперь строим интерполяцию многочленом Чебышева'''
plt.grid(True)
plt.plot(x,f(x),'k.')

p1 = che.chebinterpolate(f, 15) #Многочлен 15 степени
p2 = che.chebinterpolate(f, 20) #20
p3 = che.chebinterpolate(f, 40) #40

plt.plot(t, che.chebval(t, p1), 'r-')
plt.plot(t, che.chebval(t, p2), 'g-') 
plt.plot(t, che.chebval(t, p3), 'b-') 

plt.show()
```
