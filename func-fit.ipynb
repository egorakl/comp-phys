# Аппроксимация функции

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
def f(x, a, b, c):
    return a*x**3+b*x+c
```

```python
x1, x2 = -1, 3 #Подгонять будем на отрезке [x1, x2]
a, b, c = 4, 3, 2
n = 30 #Количество точек
```

```python
x = np.linspace(x1, x2, n)
y = f(x, a, b, c)
```

```python
'''Добавим шумов'''
np.random.seed(100)
noise = np.random.normal(0 , 20, x.size) 
ny = y + noise
```

```python
fitted, k = curve_fit(f, x, ny)
```

```python
plt.figure(figsize=(10,8))
plt.plot(x, y, 'k--', label = 'f(x)')
plt.plot(x, ny, 'ro', label = 'noise')
plt.plot(x, f(x, *fitted), 'b', label = 'fitted f(x)')
plt.legend (loc ='best' )
plt.show()
```
