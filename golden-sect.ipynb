# Нахождение min функции методом золотого сечения

```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
eps = 10 ** (-8)
k = (1 + math.sqrt(5)) / 2 #Пропорция золотого сечения
```

```python
def goldmin(a, b, f):
    while abs(b-a) > eps:
        x1 = b - (b - a) / k
        x2 = a + (b - a) / k
        y1 = f(x1)
        y2 = f(x2)
        if y1 >= y2:
            a = x1
        else:
            b = x2
    return (a + b) / 2
```

```python
'''Границы отрезка будем рисовать синими точками, найденный минимум - красной'''
def drawgld(t, a, b, f):
    plt.figure(figsize=(8,6))
    x = goldmin(a, b, f)
    print(x)
    plt.plot (t,f(t),'k', a,f(a),'bo', b,f(b),'bo', x,f(x),'ro')
    plt.show()
```

```python
'''Найдем минимум для x**2 на отрезке [-5, 5]'''
def f1(x):
    return x**2
a1 = -5.0
b1 = 5.0
```

```python
t1 = np.arange (-10, 10, 0.02)
drawgld(t1, a1, b1, f1)
```

```python
'''Теперь найдем минимум для -cos(x) на отрезке [4, 9]'''
def f2(x):
    return -np.cos(x)
a2 = 4.0
b2 = 9.0
```

```python
t2 = np.arange (0, 12, 0.02)
drawgld(t2, a2, b2, f2)
```
