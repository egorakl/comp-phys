```python
import numpy as np
import sympy
import matplotlib.pyplot as plt
import numpy.polynomial.chebyshev as che #Использовался NumPy v1.14
from mpmath import *
```

```python
def lagrange_k(xval, xk, k):
    n = len(xk)
    numer, denom = 1.0, 1.0 
    for j in range(n):
        if j == k: 
            continue
        denom *= (xk[k] - xk[j])
        numer *= (xval - xk[j])
    return numer / denom

def lagrange(xval, xk, yk):
    return sum(yk[k] * lagrange_k(xval, xk, k) for k in range(len(xk)))
```

```python
def f(x):
    return (1 / (1 + 25*x**2))
n = 15
x = np.linspace(-1, 1, n)
```

```python
t = np.linspace (-1,1,101)
l = [lagrange(i,t,f(t)) for i in t] #Интерполяция многочленом Лагранжа
```

```python
'''Рисуем интерполяцию многочленом Лагранжа'''

plt.grid(True)
plt.plot(x,f(x),'k.')

plt.plot(t,l,'m-')
plt.show()
```

```python
'''Теперь рисуем интерполяцию многочленом Чебышева'''
plt.grid(True)
plt.plot(x,f(x),'k.')

p1=che.chebinterpolate(f,15) #Многочлен 15 степени
p2=che.chebinterpolate(f,20) #20
p3=che.chebinterpolate(f,40) #40

plt.plot(t,che.chebval(t,p1),'r-')
plt.plot(t,che.chebval(t,p2),'g-') 
plt.plot(t,che.chebval(t,p3),'b-') 

plt.show()
```
