# Нахождение минимума функции методом парабол

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
eps = 10**(-8)
```

```python
def f(x):
    return 5*np.cos(4*x)+4*np.sin(8*x)
```

```python
'''Ищем минимум на отрезке [x1, x3]'''
def parab(f, x1, x3):
    while True:
        x2 = (x1 + x3) / 2
        f1 = f(x1)
        f2 = f(x2)
        f3 = f(x3)
        num = ((x2-x1)**2) * (f2-f3) - ((x2-x3)**2) * (f2-f1)
        den = 2 * ((x2-x1)*(f2-f3) - (x2-x3)*(f2-f1))
        k = num/den
        if abs(k) <= eps:
            break
        if k>=0:
            x3 = x2
        else:
            x1 = x2
    return x2
```

```python
'''Синими точками обозначим границы отрезка, красной - найденный минимум'''
def draw(f, x1, x3):
    t = np.arange (0, 3, 0.02)
    x = parab(f, x1, x3)
    plt.plot (t, f(t), 'k', x, f(x), 'ro', x1, f(x1), 'bo', x3, f(x3), 'bo')
    plt.plot ()
    plt.show()
```

```python
draw(f, 0.1, 1)
draw(f, 1.1, 2)
draw(f, 1.5, 2.5)
```
